### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```
class Tproduto:
    codigo = 0
    nome = ""
    preco = 0.0

def menu():
    print("Menu do sistema")
    print("1 - Cadastro de produtos")
    print("2 - Reajuste dos produtos")
    print("3 - Visualização dos produtos")
    print("4 - Sair")
    opcao = int(input("Qual a opção deseja: "))
    return opcao

def cadastro(vet_produto):
    for i in range(2):
        produto = Tproduto()
        produto.codigo = int(input("Cadastre o código do produto: "))
        produto.nome = input("Cadastre o nome do produto: ")
        produto.preco = float(input("Cadastre o preço do produto: R$"))
        vet_produto.append(produto)
    print("Cadastro realizado!")
    return vet_produto

def visualizar(vet_produto):
    for i in range(len(vet_produto)):
        print("Código: ", vet_produto[i].codigo, "Nome: ", vet_produto[i].nome, "Preço: ", vet_produto[i].preco)

def reajuste(vet_produto):
    for i in range(len(vet_produto)):
        vet_produto[i].preco += vet_produto[i].preco * 10 / 100
    print("Reajuste realizado com sucesso!")
    return vet_produto

def main():
    op = menu()
    vet_produto = []
    while op >= 1 and op <= 3:
        if op == 1:
            v_produto = cadastro(vet_produto)
        elif op == 2:
            v_produto = reajuste(vet_produto)
        elif op == 3:
            visualizar(vet_produto)
        op = menu()

main()
```

```
class Aluno:
    nome = ''
    data_nascimento = ''
    telefone = 0
    endereco = ''
    serie = 0


def menu():
    opcao = 0

    while opcao < 1 or opcao > 4:
        print('\nMenu de opções:\n 1. Cadastrar alunos\n 2. Consulta por nome\n 3. Visualizar todos os dados\n 4. Sair')
        opcao = int(input('Digite a opção desejada: '))

        if opcao < 1 or opcao > 4:
            print('\nOpção inválida!')

    return opcao


def cadastrar(vet_alunos):
    aluno = Aluno()

    aluno.nome = input('Cadastre o nome do aluno: ')
    aluno.data_nascimento = input('Cadastre a data de nascimento do aluno: ')
    aluno.telefone = input('Cadastre o telefone do aluno: ')
    aluno.endereco = input('Cadastre o endereço do aluno: ')
    aluno.serie = input('Cadastre a série atual do aluno: ')

    vet_alunos.append(aluno)

    return vet_alunos


def consultar(vet_alunos):
    nome = input('Digite o nome do aluno: ')

    for i in range(len(vet_alunos)):
        if vet_alunos[i].nome == nome:
            print(f'Nome do aluno: {vet_alunos[i].nome}\nData de nascimento: {vet_alunos[i].data_nascimento}\nTelefone: {vet_alunos[i].telefone}\nEndereço: {vet_alunos[i].endereco}\nSérie atual: {vet_alunos[i].serie}')
        else:
            print('Aluno não encontrado!')

def visualizar(vet_alunos):
    for i in range(len(vet_alunos)):
        print(f'Nome do aluno: {vet_alunos[i].nome}\nData de nascimento: {vet_alunos[i].data_nascimento}\nTelefone: {vet_alunos[i].telefone}\nEndereço: {vet_alunos[i].endereco}\nSérie atual: {vet_alunos[i].serie}')


def main():
    op = menu()
    vet_alunos = []

    while op >= 1 and op <= 500:
        if op == 1:
            vet_alunos = cadastrar(vet_alunos)
        elif op == 2:
            consultar(vet_alunos)
        elif op == 3:
            visualizar(vet_alunos)

        op = menu()


main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:
Menu de opções:
1. Cadastrar contas
2. Visualizar todas as contas
3. Consultar por nome
4. Alterar nome e/ou saldo de um número de conta
5. Excluir a conta com menor saldo
6. Sair

Observação:

* No item de menu 4. Alterar nome e/ou saldo de um número de conta, entenda que apenas pode ser alterado o nome e o saldo depois que você pesquisar pelo número da conta.
* No item 5 do menu, encontre o menor saldo dentre todos cadastrados, **depois exclua esta ÚNICA conta.**. O assunto de excluir algo de um vetor foi dado na disciplina de Algoritmo.



```
class classconta:
    numero: 0
    nome = ""
    saldo = 0

def menu():
    op = 0

    while op < 1 or op > 6:
        print('\nMenu de opções:\n 1. Cadastrar contas\n 2. Visualizar todas as contas\n 3. Consultar conta por nome\n 4. Alterar nome e\ou saldo de conta\n 5. Excluir conta com menor saldo\n 6. Sair')
        op = int(input('Digite a opção desejada: '))

        if op < 1 or op > 6:
            print('\nOpção inválida!')

    return op

def cadastro(vet_contas):
    for i in range(2):
        conta = classconta()
        conta.numero = int(input("\nCadastre o numero da conta: "))
        conta.nome = input('Cadastre o nome do titular: ')
        conta.saldo = float(input('Cadastre o saldo da conta: '))
        vet_contas.append(conta)

    return vet_contas

def consulta(vet_contas):
    nome = input('Digite o nome do titular: ')

    for i in range(len(vet_contas)):
        if vet_contas[i].nome == nome:
            print(f'\nNumero da conta: {vet_contas[i].numero}\nNome do titular: {vet_contas[i].nome}\nSaldo da conta: {vet_contas[i].saldo}')

def visualizacao(vet_contas):
    if len(vet_contas) > 0:
        for i in range(len(vet_contas)):
            print(f'\nNumero da conta: {vet_contas[i].numero}\nNome do titular: {vet_contas[i].nome}\nSaldo da conta: {vet_contas[i].saldo}')

def alterar(vet_contas):
    if len(vet_contas) > 0:
        numero = int(input('Digite o numero da conta: '))

        for i in range(len(vet_contas)):
            if numero == vet_contas[i].numero:
                op = input('Deseja fazer alteração do nome? (s/n)')

                if op == 's':
                    vet_contas[i].nome = input('Digite o novo nome: ')

                op = input('Deseja fazer alteração do saldo? (s/n)')

                if op == 's':
                    vet_contas[i].saldo = input('Digite o novo saldo da conta: ')

    return vet_contas

def excluir(vet_contas):
    if len(vet_contas) > 0:
        for i in range(len(vet_contas)):
            if i == 0:
                menor_saldo = vet_contas[i].saldo
                menor = i
            elif vet_contas[i].saldo < menor_saldo:
                menor_saldo = vet_contas[i].saldo
                menor = i
        vet_contas.pop(menor)
        print('Exclusão realizada com sucesso!')

    return vet_contas

def main():
    op = menu()
    vet_contas = []

    while op >= 1 and op <= 5:
        if op == 1:
            vet_contas = cadastro(vet_contas)
        elif op == 2:
            visualizacao(vet_contas)
        elif op == 3:
            consulta(vet_contas)
        elif op == 4:
            alterar(vet_contas)
        elif op == 5:
            excluir(vet_contas)

        op = menu()

main()
```

### 6. Elabore uma estrutura para representar um Funcionario (código, nome, endereço, salário). Para o membro endereço deve-se criar outra estrutura Endereço (logradouro, número, bairro, cidade). **Utilize aninhamento de estruturas** para resolver este desenvolvimento. Construa uma função para cada opçao do menu a seguir:

Menu de opções:
1. Cadastrar funcionários
2. Visualizar todos os dados
3. Sair

```
class classendereco:
    logradouro = ''
    numero = 0
    bairro = ''
    cidade = ''

class classfuncionario:
    codigo = 0
    nome = ''
    endereco = classendereco()
    salario = 0

def menu():
    op = 0

    while op < 1 or op > 3:
        print('\nMenu de opções:\n 1. Cadastrar funcionario\n 2. Visualizar todas os dados\n 3. Sair')
        op = int(input('Digite a opção desejada: '))

        if op < 1 or op > 3:
            print('\nOpção inválida!')

    return op

def cadastro(vet_funcionario):
    funcionario = classfuncionario()
    funcionario.endereco = classendereco()
    numero = 0

    while numero == 0:
        numero = int(input('\nInsira o codigo do funcionario: '))
        for i in range(len(vet_funcionario)):
            if vet_funcionario[i].codigo == numero:
                print('\nCodigo do funcionario já cadastrado!')
                numero = 0
    funcionario.codigo = numero
    funcionario.nome = input('Digite o nome do funcionario: ')
    funcionario.endereco.logradouro = input('Digite o logradouro do funcionario: ')
    funcionario.endereco.numero = int(input('Digite o numero do logradouro do funcionario: '))
    funcionario.endereco.bairro = input('Digite o bairro do funcionario: ')
    funcionario.endereco.cidade = input('Digite a cidade do funcionario: ')
    funcionario.salario = float(input('Digite o salario do funcionario: '))

    vet_funcionario.append(funcionario)

    return vet_funcionario

def visualizacao(vet_funcionario):
    for funcionario in vet_funcionario:
        print(f'\nFuncionario: {funcionario.nome}\nCodigo: {funcionario.codigo}\nEndereço: {funcionario.endereco.logradouro}, {funcionario.endereco.numero}, {funcionario.endereco.bairro}, {funcionario.endereco.cidade}\nSalario: {funcionario.salario}')

def main():
    op = menu()
    vet_funcionario = []

    while op != 3:
        if op == 1:
            cadastro(vet_funcionario)
        elif op == 2:
            visualizacao(vet_funcionario)

        op = menu()

main()
```

### 7. Elabore uma estrutura para representar um Produto (código, nome, data_fabricacao, data_validade, preço). Para o membro data_fabricacao e data_validade, deve-se criar outra estrutura Data (dia, mês, ano). **Utilize aninhamento de estruturas** para resolver este desenvolvimento. Construa uma função para cada opçao do menu a seguir:

Menu de opções:
1. Cadastrar produtos
2. Visualizar todos os dados
3. Sair

```
class Data:
    dia = 0
    mes = 0
    ano = 0

class Produto:
    codigo = 0
    nome = ''
    data_fabri = Data()
    data_vali = Data()
    preco = 0

def menu():
    print('\nMenu de opções')
    print('1. Cadastrar produtos')
    print('2. Visualizar todas os dados')
    print('3. Sair')
    op = int(input('Digite a opção desejada: '))
    return op

def cadastro(vet_produto):
    for i in range(1):
        produtos = Produto()
        produtos.codigo = int(input('Digite o codigo do produto: '))
        produtos.nome = input('Digite o nome do produto: ')
        produtos.data_fabri = Data()
        produtos.data_fabri.dia = int(input('Digite o dia de fabricação: '))
        produtos.data_fabri.mes = int(input('Digite o mes de fabricação: '))
        produtos.data_fabri.ano = int(input('Digite o ano de fabricação: '))
        produtos.data_vali = Data()
        produtos.data_vali.dia = int(input('Digite o dia de validade: '))
        produtos.data_vali.mes = int(input('Digite o mes de validade: '))
        produtos.data_vali.ano = int(input('Digite o ano de validade: '))
        produtos.preco = float(input('Digite o preço do produto: '))
        vet_produto.append(produtos)
    print('Produto cadastrado!')
    return vet_produto

def visualizar(vet_produtos):
    if len(vet_produtos) > 0:
        for i in range(len(vet_produtos)):
            print(f'\nCodigo do produto: {vet_produtos[i].codigo}\nNome do produto: {vet_produtos[i].nome}\nData de fabricação do produto: {vet_produtos[i].data_fabri.dia}/{vet_produtos[i].data_fabri.mes}/{vet_produtos[i].data_fabri.ano}\nData de validade do produto: {vet_produtos[i].data_vali.dia}/{vet_produtos[i].data_vali.mes}/{vet_produtos[i].data_vali.ano}\nPreço do produto: {vet_produtos[i].preco:.2f}')

def main():
    vet_produto = []
    op = menu()

    while op >= 1 and op <= 2:
        if op == 1:
            vet_produto = cadastro(vet_produto)
        elif op == 2:
            visualizar(vet_produto)

        op = menu()

main()
```

### 8. Elabore duas estruturas, como é apresentado a seguir:

CLIENTE |  DOCUMENTOS |
 -------|--------------
cod_cli |   num_doc   |
nome    |   cod_cli   |

        |   dia_pag   |

* Sabe-se que um documento só pode ser cadastrado para um cliente que já exista.
* Considere que podem existir, no máximo, 15 clientes e 30 documentos. Crie um vetor para clientes e outro para documentos.

* Crie um menu para a realização de cada uma das operações especificadas a seguir:

**** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS ****

1. Cadastrar clientes

2. Relatório de clientes

3. Cadastrar documentos

4. Relatório de documentos

5. Excluir clientes sem documentos

6. Excluir documentos individuais pelo número

7. Excluir documentos por cliente

8. Excluir documentos por período

9. Alterar as informações dos clientes

10. Mostrar o total de documentos de determinado cliente

11. Sair

Qual opção deseja?

.................................................................................................

Para cada item do menu, desenvolva uma função.

A seguir são apresentados os detalhes de implementação de cada opção do menu:

1. Cadastrar clientes — não pode existir mais que um cliente com o mesmo código.

2. Relatório de clientes - listar todos os clientes cadastrados.

3. Cadastrar documentos — ao cadastrar um documento, **se o dia de pagamento for maior que o dia de vencimento, calcular o campo ‘juros’ do registro documentos (5% sobre o valor original do documento)**.

4. Relatório de documentos - listar todos os documentos cadastrados.

5. Excluir clientes — um cliente só poderá ser excluído se não existir nenhum documento associado a ele.

6. Excluir documentos individuais — por meio de seu número. Caso o documento não exista, o programa deverá mostrar a mensagem "Documento não encontrado".

7. Excluir documentos por cliente — o programa deverá informar o código do cliente e excluir todos os seus documentos. Caso o cliente não exista, deverá mostrar a mensagem "Cliente não encontrado".

8. Excluir documentos por período — o programa deverá informar o dia inicial e o dia final e excluir todos os documentos que possuam data de vencimento nesse período.

9. Alterar as informações sobre os clientes — **só NÃO altere o código do cliente**.

10. Mostrar o total de documentos de determinado cliente.

```
class Clientes:
    cod_cliente = 0
    nome = ''
    fone = 0

class Documentos:
    num_doc = 0
    cod_cliente = 0
    dia_venc = 0
    dia_pg = 0
    valor = 0
    juros = 0

def menu():
    print(f'\n** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS **\n1. Cadastrar clientes\n2. Relatórios de clientes\n3. Cadastrar documentos\n4. Relatório de documentos\n5. Excluir clientes\n6. Excluir documentos individuais pelo número\n7. Excluir documentos por cliente\n8. Excluir documentos por periodo\n9. Alterar as informações dos clientes\n10. Mostrar o total de documentos de determinado cliente\n11. Sair')

    op = int(input('\nDigite a opção desejada: '))
    return op

def cadastro_clientes(vet_clientes):
    cliente = Clientes()
    numero = 0

    while numero == 0:
        numero = int(input('\nDigite o codigo do cliente: '))
        cliente.cod_cliente = numero
        cliente.nome = input('Digite o nome do cliente: ')
        cliente.fone = int(input('Digite o telefone do cliente: '))
        vet_clientes.append(cliente)

    return vet_clientes

def cadastro_documentos(vet_clientes, vet_documentos):
    documento = Documentos()
    numero = 0

    while numero == 0:
        numero = int(input('\nDigite o número do documento: '))

        for i in range(len(vet_documentos)):
            if vet_documentos[i].num_doc == numero:
                print('\nNumero de documento já cadastrado!')
                numero = 0

    documento.num_doc = numero

    existe = False

    while existe == False:
        numero = int(input('Digite o codigo do cliente: '))

        for i in range(len(vet_clientes)):
            if vet_clientes[i].cod_cliente == numero:
                documento.cod_cliente = numero
                existe = True

        if existe == False:
            print('Codigo de cliente não cadastrado!')

    documento.dia_venc = int(input('Insira o dia de vencimento do documento: '))
    documento.dia_pg = int(input('Insira o dia de pagamento do documento: '))
    documento.valor = float(input('Insira o valor do documento: '))

    if documento.dia_pg > documento.dia_venc:
        documento.juros = documento.valor * 0.05

    vet_documentos.append(documento)

    return vet_documentos

def relatorio_cliente(vet_clientes):
    print('\nRelatório de clientes\n')

    for i in range(len(vet_clientes)):
        print(f'Codigo cliente: {vet_clientes[i].cod_cliente}\nNome cliente: {vet_clientes[i].nome}\nTelefone cliente: {vet_clientes[i].fone}\n')

def relatorio_documentos(vet_documentos):
    print('\nRelatório de documentos\n')

    for i in range(len(vet_documentos)):
        print(f'Numero do documento: {vet_documentos[i].num_doc}\nCodigo do cliente: {vet_documentos[i].cod_cliente}\nDia de vencimento: {vet_documentos[i].dia_venc}\nDia de pagamento: {vet_documentos[i].dia_pg}\nValor: {vet_documentos[i].valor}\nJuros: {vet_documentos[i].juros}\n')

def excluir_clientes(vet_clientes, vet_documentos):
    numero = int(input('\nDigite o codigo do cliente: '))

    for i in range(len(vet_clientes)):
        if vet_clientes[i].cod_cliente == numero:
            for j in range(len(vet_documentos)):
                if vet_documentos[j].cod_cliente == numero:
                    print('\nCliente não pode ser excluido por possui documento (s) cadastrados!')
                    return vet_clientes

            vet_clientes.pop(i)
            print('\nCliente excluido com sucesso!')
            break

    return vet_clientes

def excluir_documentos_ind(vet_documentos):
    excluir = False
    numero = int(input('\nDigite o numero do documentos: '))

    for i in range(len(vet_documentos)):
        if vet_documentos[i].num_doc == numero:
            vet_documentos.pop(i)
            print('\nDocumento excluido com sucesso!')
            excluir = True

    if excluir == False:
        print('\nDocumento não encontrado!')

    return vet_documentos

def excluir_documentos_cliente(vet_clientes, vet_documentos):
    numero = int(input('\nDigite o numero do documentos: '))
    existe = False

    for i in range(len(vet_clientes)):
        if vet_clientes[i].cod_cliente == numero:
            existe = True

    posi_documentos = []

    if existe == True:
        for i in range(len(vet_documentos)):
            if vet_documentos[i].cod_cliente == numero:
                posi_documentos.append(i)

    else:
        print('\nCliente não encontrado!')

    for i in range(len(posi_documentos)):
        vet_documentos.pop(posi_documentos[i])
        print('\nDocumento excluido com sucesso!')

    return vet_documentos

def excluir_documentos_periodo(vet_documentos):
    data_inicial = int(input('\nDigite a data inicial: '))
    data_final = int(input('\nDigite a data final: '))

    for i in range(len(vet_documentos)):
        if vet_documentos[i].dia_pg >= data_inicial and vet_documentos[i].dia_pg <= data_final:
            vet_documentos.pop(i)
            print('\nDocumentos excluido com sucesso!')

    return vet_documentos

def alterar_cliente(vet_clientes):
    numero = int(input('\nDigite o codigo do cliente: '))

    for i in range(len(vet_clientes)):
        if vet_clientes[i].cod_cliente == numero:
            op = (input('\nDeseja alterar o nome do cliente? (s/n): '))
            if op == 'S' or op == 's':
                vet_clientes[i].nome = input('\nDigite o novo nome do cliente: ')
            op = (input('\nDeseja alterar o telefone do cliente? (s/n): '))
            if op == 'S' or op == 's':
                vet_clientes[i].fone = input('\nDigite o novo telefone do cliente: ')
            print('\nCliente alterado com sucesso!')
            break

    return vet_clientes

def totalizar_documentos(vet_documentos):
    total = 0
    numero = int(input('\nDigite o codigo do cliente: '))

    for i in range(len(vet_documentos)):
        if vet_documentos[i].cod_cliente == numero:
            total += 1
    print(f'\nTotal de documentos: {total}')

def main():
    vet_clientes = []
    vet_documentos = []
    op = menu()

    while op >= 1 and op <= 10:
        if op == 1:
            vet_clientes = cadastro_clientes(vet_clientes)
        elif op == 2:
            relatorio_cliente(vet_clientes)
        elif op == 3:
            vet_documentos = cadastro_documentos(vet_clientes, vet_documentos)
        elif op == 4:
            relatorio_documentos(vet_documentos)
        elif op == 5:
            vet_clientes = excluir_clientes(vet_clientes, vet_documentos)
        elif op == 6:
            vet_documentos = excluir_documentos_ind(vet_documentos)
        elif op == 7:
            vet_documentos = excluir_documentos_cliente(vet_clientes, vet_documentos)
        elif op == 8:
            vet_documentos = excluir_documentos_periodo(vet_documentos)
        elif op == 9:
            vet_clientes = alterar_cliente(vet_clientes)
        elif op == 10:
            totalizar_documentos(vet_documentos)

        op = menu()

main()
```
